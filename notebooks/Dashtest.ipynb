```python
import dash_bootstrap_components as dbc
from dash import Dash, html, dcc, Input, Output
from dash import dash_table
import plotly.express as px

app = Dash(__name__, external_stylesheets=[dbc.themes.PULSE])
df = px.data.iris()

app.layout = html.Div([
    dcc.Dropdown(
        id="species",
        options=[{"label": s, "value": s} for s in df["species"].unique()],
        value="versicolor"
    ),
    dcc.Graph(id="graph"),
    dash_table.DataTable(
        data=df.to_dict("records"),
        columns=[{"name": i, "id": i} for i in df.columns],
        style_table={"overflowX": "auto"},
        style_cell={"textAlign": "left"},
    )
])

@app.callback(
    Output("graph", "figure"),
    Input("species", "value")
)
def update_graph(species):
    filtered = df[df["species"] == species]
    fig = px.scatter(filtered, x="sepal_width", y="sepal_length")
    fig.update_layout(
        paper_bgcolor='rgb(30,30,30)',
        plot_bgcolor='rgb(30,30,30)',
        font_color='white'
    )
    return fig

app.run(jupyter_mode="external")  # or use JupyterDash if needed
```
